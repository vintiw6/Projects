```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv' , encoding = 'unicode_escape')
# to avoid encoding error, we use 'unicode escape'
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
# To check for null values we use this as it gives us the total sum of null values
pd.isnull(df).sum()
```

```python
# To drop null values we use dropna
df.dropna(inplace=True)
```

```python
# To initialize list of lists
data_test = [['madhav', 11], ['Gopi', 15], ['Keshav'], ['Lalita', 16]]

# To Create the pandas DataFrame using the list
df_test = pd.DataFrame(data_test, columns=['Name', 'Age'])

# To display the DataFrame
df_test
```

```python
# To change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# To rename any column
df.rename(columns = {'Martial_status':'Shaadi'})
```

```python
# describe() method returns description of the data in the DataFrame(i.e. count, mean, std, etc)
df.describe()
```

```python
# To use describe() for specific coulmns
df[['Age', 'Orders', 'Amount']].describe()
```

**_EDA (Exploritary Data Analysis)_**

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

**From the Above graph we can see most of the buyers are female and hence the purchasing power is more for the female in comparison to a male**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
# hue lets us see the different categories in the same age group as male and female

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount Vs Age Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age)
```

**From The Above graph we can see that most of the buyers are from age group 26-35 followed by 36-45 yrs female**

**_STATE_**

```python
# Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state, x = 'State' , y = 'Orders')
```

```python
# Total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state, x = 'State' , y = 'Amount')
```

**From the above we can see that most of the orders are from Uttar Pradesh,Maharastra and Karnataka respectively**

**_Marital Status_**

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(20,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_marital_status, x = 'Marital_Status' , y = 'Amount' , hue = 'Gender')
```

**From the above we can see that most of the buyers are Married(women) and they have the highest purchasing power**

**_Occupation_**

```python
ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(20,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_occupation, x = 'Occupation' , y = 'Amount')
```

**From above graph we can see the most of the buyers are working in IT, Healthcare, Aviation**

**_Product Category_**

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(30,20)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_product_category, x = 'Product_Category' , y = 'Amount')
```

**From above graph we can see the most sold Products belongs to Food, Footwear, and Electronics Categories**

```python
sales_Product_ID = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_Product_ID, x = 'Product_ID' , y = 'Orders')
```

**_Conclusion_**

_Married women from age group 26-35 yrs from UP, Maharastra, Karnataka working in IT, Aviation, and Healthcare are more likely to buy product from food, clothing and Electronics Category_
